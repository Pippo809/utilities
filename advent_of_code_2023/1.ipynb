```python
import pandas as pd
```

```python
values = []
with open(r"advent_of_code_2023\1.txt", "+r") as file:
    values = file.readlines()
```

```python
tot = 0
for line in values:
    first = None
    last = None
    idx = 0
    while not first:
        d = line[idx]
        if d.isdigit():
            first = d
        idx += 1
    idx = len(line)-1
    while not last:
        d = line[idx]
        if d.isdigit():
            last = d
        idx -= 1
    tot += int(f"{first}{last}")
```

```python
val_digits = {"one": 1, "two": 2, "three":3, "four":4, "five":5, "six":6, "seven":7, "eight":8, "nine":9}
```

```python
tot = 0
for line in values:
    line.rstrip("\n")
    print(line)
    first = None
    last = None
    idx = 0
    while not first:
        d = line[idx]
        written = line[:idx+1]
        for k, v in val_digits.items():
            if k in written:
                first = v
                break
        if d.isdigit():
            first = d
        idx += 1
    idx = len(line)-1
    while not last:
        d = line[idx]
        written = line[idx:]
        if d.isdigit():
            last = d
        for k, v in val_digits.items():
            if k in written:
                last = v
                break
        idx -= 1
    # print(line, first, last)
    tot += int(f"{first}{last}")
print(tot)
```

```python
def part1(data):
    return sum((digits := [int(i) for i in line if i.isdigit()])[0] * 10 + digits[-1] for line in data)
```

```python
def part2(data):
    mappings = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
    new_data = [[x if (x := "".join([str(idx) for idx, val in enumerate(mappings, 1) if line[i:].startswith(val)])) else line[i] for i in range(len(line))] for line in data]
    return part1(new_data)
```

```python
part2(values)
```
