```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file = r'd:\OneDrive\OneDrive - uniroma1.it\Università\Tesi Master\plots\Racetrack\exp_data\16_acts_vpp_from_area_2\experiment_data_02_21-14-17.csv'
df = pd.read_csv(csv_file)
df.head()
```

```python
df = df[df["state"] != 0]
df = df[df["state"] != 1]
```

```python
diff = df["state"].diff()
df["episode"] = (diff < 0).cumsum()
```

```python
for episode in range(0, df["episode"].max()):
    episode_df = df[df["episode"] == episode]
    dist_x = episode_df["agent_location_x"].diff()
    dist_y = episode_df["agent_location_y"].diff()
    dist = np.sqrt(dist_x**2 + dist_y**2)
    df.loc[df["episode"] == episode, "distance"] = dist
```

```python
fps = 12
df["speed"] = df["distance"] * fps
df.to_csv(r'F:\images_paper\CSVs\experiment_data_02_21-14-17_speed.csv')
```
