# Birthdays paradox

n people in a room, what is the probability that at least two of them have the same birthday?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Person():
    def __init__(self) -> None:
        self.birthday = np.random.choice([1, 365])
```

```python
def has_duplicates(lst):
    seen = set()
    for i in lst:
        if i in seen:
            return True
        seen.add(i)
    return False
```

```python
def has_duplicates_birtdays(n):    
    trials = int(1e3)
    dupl = 0

    for _ in range(trials):
        people = (np.random.choice(364) for _ in range(n))
        if has_duplicates(people): dupl += 1

    expected_val = dupl/trials
    return expected_val
```

```python
print(has_duplicates_birtdays(23))
```

```python
x = np.arange(100)
y = [has_duplicates_birtdays(x[i]) for i in x]
plt.plot(x, y)
```
